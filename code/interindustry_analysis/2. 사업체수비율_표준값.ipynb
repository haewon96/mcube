# 경기지역화폐 가맹점 지역별 업종별 사업체수비율 ----- (1)

## 사업체수비율 표준값 분자 = (지역별 업종별 사업체수 / 경기지역전체 업종별 사업체수)

```python
업종별 = ggstore_region_t.index
업종별
```

```python
지역별 = ggstore_region_t.columns
지역별
```

```python
사업체수비율 = DataFrame(index = 업종별, columns = 지역별)

for region in 지역별 :
    사업체수비율[region] = DataFrame(round(ggstore_region_t[region] / 업종별사업체총수, 4))
```

```python
사업체수비율
```

# 경기지역화폐 가맹점 지역별 업종별 사업체수비율 최대값 ----- (2)

## 사업체수비율 표준값 분모 = 지역별 업종별 사업체수비율 최대값

```python
사업체수비율_최대값 = 사업체수비율.max(axis = 1)
사업체수비율_최대값
```

```python
사업체수비율_최대값_df = DataFrame(사업체수비율_최대값)
사업체수비율_최대값_df.columns = ['사업체수비율_최대값']
사업체수비율_최대값_df
```

# 경기지역화폐 가맹점 지역별 업종별 사업체수비율 표준값 ----- (3)

## (1) / (2)

```python
사업체수비율_표준값 = DataFrame(index = 업종별, columns = 지역별)

for region in 지역별 :
    사업체수비율_표준값[region] = DataFrame(round(round(ggstore_region_t[region] / 업종별사업체총수, 4) / 사업체수비율_최대값, 4))
```
