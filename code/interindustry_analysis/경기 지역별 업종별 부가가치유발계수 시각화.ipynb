## R ggplot

```python
# 역시 plot은 R이 최고야
from warnings import filterwarnings
from plotnine import *
filterwarnings('ignore')

import matplotlib.font_manager as fm
fontpath = 'C:/Users/User/AppData/Local/Microsoft/Windows/Fonts/NanumBarunGothic.ttf'
font = fm.FontProperties(fname=fontpath, size=9)
```

```python
tidy_table = vaic_stn_t4.reset_index().melt('index')
```

```python
tidy_table.columns = ['품목', '지역','부가가치유발계수']
```

```python
(ggplot(tidy_table, aes(x='품목',y='부가가치유발계수', fill ='품목'))
 + geom_col(stat='identity', position='dodge')
 + labs(x='', y='부가가치유발계수', title='4개 지역별 업종별 부가가치유발계수')
 + facet_wrap('~지역')
 + theme(text=element_text(fontproperties=font),
        axis_text_x = element_text(angle = 90,  size=0.01))
)
```
