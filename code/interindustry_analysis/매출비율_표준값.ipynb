# 경기지역화폐 지역별 업종별 매출액 

```python
DataFrame(gglc.groupby(['결제상품명', '가맹점업종명'])['결제총액'].sum())
```

```python
gglc_region = gglc.pivot_table(index = '결제상품명', 
                               columns = '가맹점업종명',
                               aggfunc = 'sum',
                               values = '결제총액').fillna(0)
gglc_region
```

```python
gglc_region.info()
```

```python
gglc_region.index
```

```python
gglc_region.index = ['남양주', '가평', '고양', '과천', '광명', '광주', '구리', '군포', '동두천', '부천',
                    '수원', '안산', '안성', '양주', '양평', '여주', '연천', '오산', '용인', '의정부',
                    '이천', '파주', '하남', '화성']
gglc_region
```

```python
gglc_region_t = gglc_region.transpose()
gglc_region_t
```

```python
gglc_region_t.index
```

```python
gglc_region_t.columns
```

# 경기지역화폐 지역별 업종별 매출총액 

```python
지역별매출총액 = gglc_region_t.sum(axis = 0)
지역별매출총액
```

```python
지역별매출총액df = DataFrame(지역별매출총액)
지역별매출총액df.columns = ['지역별매출총액']
지역별매출총액df
```

```python
업종별매출총액 = gglc_region_t.sum(axis = 1)
업종별매출총액
```

```python
업종별매출총액df = DataFrame(업종별매출총액)
업종별매출총액df.columns = ['업종별매출총액']
업종별매출총액df
```

# 경기지역화폐 지역별 업종별 매출비율     -----     (1)

## 매출비율 표준값 분자 = (지역별 업종별 매출 / 경기지역전체 업종별 매출)

```python
매출비율 = DataFrame(index = gglc_region_t.index, columns = gglc_region_t.columns)

for region in gglc_region_t.columns :
    매출비율[region] = DataFrame(round(gglc_region_t[region] / 업종별매출총액, 4))
```

```python
매출비율
```

# 경기지역화폐 지역별 업종별 매출비율 최대값     -----     (2)

## 매출비율 표준값 분모 = 지역별 업종별 매출비율 최대값

```python
매출비율_최대값 = 매출비율.max(axis = 1)
매출비율_최대값
```

```python
매출비율_최대값_df = DataFrame(매출비율_최대값)
매출비율_최대값_df.columns = ['매출비율_최대값']
매출비율_최대값_df
```

# 경기지역화폐 지역별 업종별 매출비율 표준값     -----     (3)

## (1) / (2)

```python
매출비율_표준값 = DataFrame(index = gglc_region_t.index, columns = gglc_region_t.columns)

for region in gglc_region_t.columns :
    매출비율_표준값[region] = DataFrame(round(round(gglc_region_t[region] / 업종별매출총액, 4) / 매출비율_최대값, 4))
```
