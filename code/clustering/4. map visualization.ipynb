```python
import pandas as pd
import folium
import json
```

## 경기도 클러스터링 시각화

```python
pdf = pd.read_csv('../../data/clustering_data/경기클러스터.csv')
```

```python
pdf['cluster_id'] = pdf['cluster_id'].astype('object')
```

```python
pdf
```

```python
geo_data = '../../data/clustering_data/TL_SCCO_SIG.json'
geo_str = json.load(open(geo_data, encoding='utf-8'))
```

1) 임의로 구 추가
#구가 있는 시는 'SIG_KOR_NM': '성남시 중원구'
#고양시 : 일산동구 일산서구 덕양구
#성남시 : 중원구 분당구 수정구
#수원시 : 장안구 권선구 팔달구 영통구
#안산시 : 상록구 단원구
#용인시 : 수지구 기흥구 처인구

```python
gu = ['고양시 일산동구', '고양시 일산서구', '고양시 덕양구' '성남시 중원구', '성남시 분당구', '성남시 수정구', '수원시 장안구',
     '수원시 권선구', '수원시 팔달구', '수원시 영통구', '안산시 상록구', '안산시 단원구', '용인시 수지구', '용인시 기흥구, 용인시 처인구']
```

```python
gu_pdf = pdf
```

```python
for g in gu:
    si = g[:3]
    new = pdf[pdf['시군'] == si]
    new['시군'] = g 
    gu_pdf = pd.concat([gu_pdf, pd.DataFrame(new)],ignore_index=True)
    #gu_pdf.append(pd.DataFrame(new), ignore_index=True)
    #gu_pdf.append(gu_pdf.iloc[-1], ignore_index=True)
    #gu_pdf.iloc[-1] = list(new)
```

```python
gu_pdf
```

2) 시군구 코드 추가

```python
cd = pd.read_excel('../../data/clustering_data/법정동코드 조회자료.xls')
```

```python
cd.head()
```

```python
cd['법정동코드'] = cd['법정동코드'].apply(lambda x : str(str(x)[:5]))
```

```python
gu_pdf['시군'] = gu_pdf['시군'].apply(lambda x : '경기도 '+ x)
```

```python
cd.rename(columns = {'법정동코드' : '시군코드', '법정동명':'시군'}, inplace = True)
```

```python
df = pd.merge(gu_pdf, cd, how='left', on='시군')
```

```python
df
```

### 지도

```python
center = [37.541, 126.986]

# 맵이 center 에 위치하고, zoom 레벨은 9로 시작하는 맵 m
m = folium.Map(location=center, zoom_start=8.5)

# Choropleth 레이어를 만들고, 맵' m에 추가
folium.Choropleth(
    geo_data=geo_str,
    data=df,
    columns=('시군코드', 'cluster_id'),
    key_on='feature.properties.SIG_CD',
    fill_color='Set3',
    threshold_scale=[0, 1, 2, 3, 4, 5, 6],
    legend_name='cluster'
).add_to(m)

```
