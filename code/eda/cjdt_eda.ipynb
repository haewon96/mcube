```python
#0817회의 후 역할분담
#cj물류데이터 EDA 정아

import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime
```

```python
#제공데이터 불러오기

cjdt = pd.read_excel \
('basicdata/05_물류데이터(CJ올리브네트웍스)/2020 빅콘테스트_CJ올리브네트웍스_제공DB.xlsx')
```

```python
cjdt['DL_YMD'] = pd.to_datetime(cjdt['DL_YMD'], format='%y%m%d')
cjdt.info()
cjdt.head()
```

```python
cjdt['DL_YM'] = cjdt.DL_YMD.apply(lambda X: X.strftime('%y%m'))
```

```python
cjdt['DL_Y'] = cjdt.DL_YMD.apply(lambda X: X.strftime('%Y'))
```
