# 불러오기

```python
bedtown_path = '/../../data/clustering_data/침상도시/'
bedtown_list = os.listdir(bedtown_path)
bedtown_list
```

```python
#불러오기
residence_Daegu = pd.read_csv(bedtown_path+'거주형태_대구.csv', encoding = 'CP949', header = 1)
residence_Seoul = pd.read_csv(bedtown_path+'거주형태_서울.csv', encoding = 'CP949', header = 1)
residence_Gyeongi = pd.read_csv(bedtown_path+'거주형태_경기.csv', encoding = 'CP949', header = 1)

demography_Daegu = pd.read_csv(bedtown_path+'인구통계_대구.csv', encoding = 'CP949', header = 1)
demography_Seoul = pd.read_csv(bedtown_path+'인구통계_서울.csv', encoding = 'CP949').drop(['Unnamed: 6','단위'], axis = 1)
demography_Gyeongi = pd.read_csv(bedtown_path+'인구통계_경기.csv', encoding = 'CP949').drop(['Unnamed: 6','단위'], axis = 1)
```

```python
#거주형태
residence_주택_Daegu = residence_Daegu[residence_Daegu.거처의종류 == '주택'][['행정구역별(시군구)','거처']]
residence_주택_Seoul = residence_Seoul[residence_Seoul.거처의종류 == '주택'][['행정구역별(시군구)','거처']]
residence_주택_Gyeongi = residence_Gyeongi[residence_Gyeongi.거처의종류 == '주택'][['행정구역별(시군구)','거처']]

residence_주택_Daegu.columns = ['행정구역별(시군구)', '주택']
residence_주택_Seoul.columns = ['행정구역별(시군구)', '주택']
residence_주택_Gyeongi.columns = ['행정구역별(시군구)', '주택']
```

```python
residence_주택_Daegu['전체대비주택비율'] = residence_주택_Daegu.주택/759619 #대구전체 주택수
residence_주택_Seoul['전체대비주택비율'] = residence_주택_Seoul.주택/2860562
residence_주택_Gyeongi['전체대비주택비율'] = residence_주택_Gyeongi.주택/4075961
```

```python
area_Daegu = pd.read_csv('/../../data/clustering_data/대구_e지표_정리.csv').drop(['Unnamed: 0'], axis = 1)[['지역별','도시지역면적']]
area_Seoul = pd.read_csv('/../../data/clustering_data/서울_e지표_정리.csv').drop(['Unnamed: 0'], axis = 1)[['지역별','도시지역면적']]
area_Gyeongi = pd.read_csv('/../../data/clustering_data/경기_e지표_정리.csv').drop(['Unnamed: 0'], axis = 1)[['지역별','도시지역면적']]

area_Daegu.columns = ['행정구역별(시군구)','도시지역면적']
area_Seoul.columns = ['행정구역별(시군구)','도시지역면적']
area_Gyeongi.columns = ['행정구역별(시군구)','도시지역면적']

residence_주택_Daegu = pd.merge(area_Daegu,residence_주택_Daegu)
residence_주택_Seoul = pd.merge(area_Seoul,residence_주택_Seoul)
residence_주택_Gyeongi = pd.merge(area_Gyeongi,residence_주택_Gyeongi)

residence_주택_Daegu['면적대비주택비율'] = residence_주택_Daegu.apply(lambda x : (x['주택']/ x['도시지역면적']), axis = 1)
residence_주택_Seoul['면적대비주택비율'] = residence_주택_Seoul.apply(lambda x : (x['주택']/ x['도시지역면적']), axis = 1)
residence_주택_Gyeongi['면적대비주택비율'] = residence_주택_Gyeongi.apply(lambda x : (x['주택']/ x['도시지역면적']), axis = 1)
```

```python
#인구통계
df1 = demography_Daegu[demography_Daegu.성별 == '계'][['행정구역별(시군구)','연령별','일반가구원']]

demography_Seoul = demography_Seoul[demography_Seoul.성별 == '계']
df2 = demography_Seoul[demography_Seoul.항목 == '일반가구원'][['행정구역별(시군구)','연령별','2019 년']]

demography_Gyeongi = demography_Gyeongi[demography_Gyeongi.성별 == '계']
df3 = demography_Gyeongi[demography_Gyeongi.항목 == '일반가구원'][['행정구역별(시군구)','연령별','2019 년']]
```

```python
demography_가구원_Daegu = df1.set_index(['연령별','행정구역별(시군구)']).unstack('연령별').reset_index()
demography_가구원_Seoul = df2.set_index(['연령별','행정구역별(시군구)']).unstack('연령별').reset_index()
demography_가구원_Gyeongi = df3.set_index(['연령별','행정구역별(시군구)']).unstack('연령별').reset_index()

demography_가구원_Daegu.columns = demography_가구원_Daegu.columns.droplevel([0])
demography_가구원_Seoul.columns = demography_가구원_Seoul.columns.droplevel([0])
demography_가구원_Gyeongi.columns = demography_가구원_Gyeongi.columns.droplevel([0])

demography_가구원_Daegu.columns = ['행정구역별(시군구)', '15~19세', '15세미만', '20~24세', '25~29세', '30~34세', '35~39세', '40~44세',
       '45~49세', '50~54세', '55~59세', '60~64세', '65~69세', '70~74세', '75~79세',
       '80~84세', '85세이상', '합계']
demography_가구원_Seoul.columns = ['행정구역별(시군구)', '15~19세', '15세미만', '20~24세', '25~29세', '30~34세', '35~39세', '40~44세',
       '45~49세', '50~54세', '55~59세', '60~64세', '65~69세', '70~74세', '75~79세',
       '80~84세', '85세이상', '합계']
demography_가구원_Gyeongi.columns = ['행정구역별(시군구)', '15~19세', '15세미만', '20~24세', '25~29세', '30~34세', '35~39세', '40~44세',
       '45~49세', '50~54세', '55~59세', '60~64세', '65~69세', '70~74세', '75~79세',
       '80~84세', '85세이상', '합계']
```

```python
df_Daegu = pd.merge(residence_주택_Daegu[:],demography_가구원_Daegu)
df_Seoul = pd.merge(residence_주택_Seoul[:],demography_가구원_Seoul)
df_Gyeongi = pd.merge(residence_주택_Gyeongi[:],demography_가구원_Gyeongi)

df_Seoul.iloc[:,1:] = df_Seoul.iloc[:,1:].apply(pd.to_numeric)
df_Gyeongi.iloc[:,1:] = df_Gyeongi.iloc[:,1:].apply(pd.to_numeric)
```

```python
#df_Seoul.info()
#df_Gyeongi.info()
```

```python
df_list = [df_Daegu,df_Seoul,df_Gyeongi]
for df in df_list:
    #print(df)
    df['미성년'] = df['15~19세']+df['15세미만']
    df['청년'] = df['20~24세']+df['25~29세']+df['30~34세']
    df['중년'] = df['35~39세']+df['40~44세']+df['45~49세']
    df['장년'] = df['50~54세']+df['55~59세']+df['60~64세']
    df['노년'] = df['65~69세']+df['70~74세']+df['75~79세']+df['80~84세']+df['85세이상']
    
    df['미성년비율'] = df.apply(lambda x : x['미성년']/ x['합계'], axis = 1)
    df['청년비율'] = df.apply(lambda x : x['청년']/ x['합계'], axis = 1)
    df['중년비율'] = df.apply(lambda x : x['중년']/ x['합계'], axis = 1)
    df['장년비율'] = df.apply(lambda x : x['장년']/ x['합계'], axis = 1)
    df['노년비율'] = df.apply(lambda x : x['노년']/ x['합계'], axis = 1)
```

```python
df_Daegu = df_Daegu[['행정구역별(시군구)', '주택', '전체대비주택비율','면적대비주택비율','미성년비율', '청년비율', '중년비율', '장년비율', '노년비율']]
df_Seoul = df_Seoul[['행정구역별(시군구)', '주택', '전체대비주택비율','면적대비주택비율','미성년비율', '청년비율', '중년비율', '장년비율', '노년비율']]
df_Gyeongi = df_Gyeongi[['행정구역별(시군구)', '주택', '전체대비주택비율','면적대비주택비율','미성년비율', '청년비율', '중년비율', '장년비율', '노년비율']]
```

```python
df_Daegu.to_csv("대구_주택과연령비율.csv", mode='w')
df_Seoul.to_csv("서울_주택과연령비율.csv", mode='w')
df_Gyeongi.to_csv("경기_주택과연령비율.csv", mode='w')
```
