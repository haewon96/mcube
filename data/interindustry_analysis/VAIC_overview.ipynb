```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'AppleGothic'
from matplotlib import pyplot
```

```python
vaic_stn = pd.read_excel('부가가치유발계수_표준_업종.xlsx', \
                       sheet_name='산업연관표_표준행렬_부가가치유발계수')
vaic_stn
```

```python
vaic_stn=vaic_stn.set_index('Unnamed: 0')
```

```python
import seaborn as sns

sns.set(font="AppleGothic", 
        rc={"axes.unicode_minus":False},
        style='darkgrid')
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(data=vaic_stn, annot=False, fmt='.2f', \
            linewidths=.5, cmap='Blues')
plt.title('표준 부가가치유발계수 업종별 정리', fontsize=20)
```
